#### Interface screen development

```python
import tkinter as tk
from tkinter import *
from tkinter.ttk import Frame,Style
```

```python
msg ="Welcome to KAN CHATBOT (Rev. A00)"
root = tk.Tk()
root.title(msg)
style = Style()

def printinfo():
    print("Input question: ",labelE.get())
  
#Create chatbot
label1=Label(root, text="Chatbot",font=('Times', 24))
label1.grid(row=0,column=0)
label2=Label(root, text="Input Question: ",font=('Times', 13))
label2.grid(row=10, column=0)

labelE=Entry(root,font=("arial", 20))
labelE.grid(row=11,column=0, padx=10)
chatbtn=Button(root,text="Start_Chat",command=printinfo)
chatbtn.grid(row=12,column=0)

root.mainloop()
```
